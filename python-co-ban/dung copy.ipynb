```python
import pygame
import sys

pygame.init()

screen = pygame.display.set_mode((800, 600))
pygame.display.set_caption('Di chuyển hình vuông')

white = (255, 255, 255)
black = (0, 0, 0)

square_x = 100
square_y = 100
square_size = 50
velocity = 5

clock = pygame.time.Clock()
fps = 60

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT]:
        square_x -= velocity
    if keys[pygame.K_RIGHT]:
        square_x += velocity
    if keys[pygame.K_UP]:
        square_y -= velocity
    if keys[pygame.K_DOWN]:
        square_y += velocity

    screen.fill(white)
    pygame.draw.rect(screen, black, (square_x, square_y, square_size, square_size))
    pygame.display.flip()
    clock.tick(fps)

pygame.quit()
sys.exit()
```

```python
import socket

# Tạo socket client
client_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
client_socket.connect(('localhost', 12345))  # Kết nối tới server tại địa chỉ IP và cổng 12345

print("Đã kết nối tới server")

# Gửi dữ liệu tới server
while True:
    message = input("Nhập tin nhắn gửi cho server: ")
    client_socket.send(message.encode())
    # Nhận phản hồi từ server
    data = client_socket.recv(1024)
    if not data:
        break
    print("Server:", data.decode())

# Đóng kết nối
client_socket.close()
```

```python
import cv2
import numpy as np

# Khởi tạo camera (nếu sử dụng webcam)
cap = cv2.VideoCapture(0)

while True:
    # Đọc một khung hình từ camera
    ret, frame = cap.read()
    if not ret:
        break

    # Chuyển đổi từ BGR sang không gian màu HSV
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # Định nghĩa ngưỡng màu đỏ trong không gian HSV
    # Màu đỏ thường có hai dải, do nằm ở đầu và cuối của thang HSV
    lower_red1 = np.array([0, 120, 70])
    upper_red1 = np.array([10, 255, 255])
    lower_red2 = np.array([170, 120, 70])
    upper_red2 = np.array([180, 255, 255])

    # Tạo mask để chỉ giữ lại các vùng có màu đỏ
    mask1 = cv2.inRange(hsv, lower_red1, upper_red1)
    mask2 = cv2.inRange(hsv, lower_red2, upper_red2)
    mask = mask1 + mask2

    # Lọc và tìm các khung hình vuông bao quanh các đối tượng có màu đỏ
    contours, _ = cv2.findContours(mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    for contour in contours:
        area = cv2.contourArea(contour)
        # Chỉ vẽ khung nếu vùng màu đỏ đủ lớn (giảm nhiễu)
        if area > 500:
            x, y, w, h = cv2.boundingRect(contour)
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
            cv2.putText(frame, "red", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    # Hiển thị khung hình với kết quả
    cv2.imshow("Red Detection", frame)

    # Thoát khi nhấn phím 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Giải phóng camera và đóng cửa sổ
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import mediapipe as mp

# Khởi tạo Mediapipe Hands
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
hands = mp_hands.Hands(min_detection_confidence=0.7, min_tracking_confidence=0.7)

# Khởi tạo camera (webcam)
cap = cv2.VideoCapture(0)

# Hàm để đếm số ngón tay được giơ lên
def count_fingers(landmarks):
    # Thứ tự các điểm: [0, 1, 2, 3, ..., 20] (0 là cổ tay, 4 là đầu ngón cái)
    finger_tips_ids = [4, 8, 12, 16, 20]

    fingers = []

    # Kiểm tra ngón cái
    if landmarks[finger_tips_ids[0]].x < landmarks[finger_tips_ids[0] - 1].x:
        fingers.append(1)  # Ngón cái giơ lên
    else:
        fingers.append(0)

    # Kiểm tra các ngón còn lại
    for tip_id in finger_tips_ids[1:]:
        if landmarks[tip_id].y < landmarks[tip_id - 2].y:
            fingers.append(1)  # Ngón giơ lên
        else:
            fingers.append(0)

    return fingers.count(1)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Lật ngược khung hình để giống với gương
    frame = cv2.flip(frame, 1)
    h, w, c = frame.shape

    # Chuyển đổi từ BGR sang RGB
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    result = hands.process(rgb_frame)

    if result.multi_hand_landmarks:
        for hand_landmarks in result.multi_hand_landmarks:
            # Vẽ khung bàn tay
            mp_drawing.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)

            # Lấy các tọa độ landmark
            landmarks = hand_landmarks.landmark

            # Đếm số ngón tay được giơ lên
            fingers_up = count_fingers(landmarks)

            # Hiển thị số ngón tay được đếm
            cv2.putText(frame, f'Fingers: {fingers_up}', (10, 50),
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    # Hiển thị khung hình
    cv2.imshow("Finger Counting", frame)

    # Thoát khi nhấn phím 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Giải phóng camera và đóng cửa sổ
cap.release()
cv2.destroyAllWindows()
```

Như vậy, trong bài này chúng ta đã tìm hiểu được về cách thức thực hiện một số thao tác trên tập dữ liệu Tuple. Về cơ bản chúng ta có thể rút ra một số điều sau đây về kiểu dữ liệu tuple:

Tuple là kiểu dữ liệu bất biến do đó các phép xóa, sửa, thêm phần từ vào tuple là không thực hiện được. Chúng ta chỉ có thể thực hiện được khi ép kiểu dữ liệu sang dạng danh sách List sau đó thực hiện các thao tác này trên đó và ép ngược trở lại thành tuple
Các đối tượng kiểu Tuple và List có thể được lồng bên trong một tập hợp kiểu Tuple. Chúng ta có thể thực hiện truy cập vào các đối tượng kiểu Tuple được lồng bên trong bằng cách sử dụng chỉ số thông thường.
Các phần tử được lồng bên trong đối tượng kiểu Tuple có thể được thay đổi nếu chúng thuộc kiểu dữ liệu có thể thay đổi được như List.
Các thao tác khác trên Tuple về cơ bản tương tự như các thao tác trên List ngoại trừ các hàm thêm, sửa, xóa như đã nói ở trên.
Trong bài tiếp theo chúng ta sẽ cùng luyện tập với một số bài tập về các thao tác cơ bản đã học trên tuple.

```python
# Giả sử items là một tập hợp (set) và lit là một danh sách chứa các tập hợp
items = {1, 2, 3, 4}
lit = [{1, 2}, {2, 3, 4}, {4, 5}, {1, 3}]

n = len(lit)
x = 0

while n:
    intersection = items & lit[n-1]  # Lấy giao của 2 tập hợp
    print(f"Processing set {n}: {lit[n-1]}, Intersection: {intersection} -> ",
          "Match" if len(intersection) > len(lit[n-1]) * 4 / 10 else "No Match")
    # Tăng biến đếm nếu điều kiện đúng
    if len(intersection) > len(lit[n-1]) * 4 / 10:
        x += 1
    n -= 1

print("Total matches:", x)
```

```python
text = input('chuỗi cần mã hóa') 
k = int(input()) 

str_result = ""

for char in text:
  if ord(char) >= 97 and ord(char) <= 122:
    str_result = str_result + chr((ord(char) - 97 + k) % 26 + 97)
  elif ord(char) >= 65 and ord(char) <= 90:
    str_result = str_result + chr((ord(char) - 65 + k) % 26 + 65)
  else:
    str_result = str_result + char

print(f'sau khi chuẩn hóa: {str_result}')
```

Câu 2. (2 điểm): Để quản lý hàng hoá cho một cửa hàng tạp hoá, người ta sử
dụng một từ điển gồm các cặp key-value sau:
• key_1 là “Ten” - value_1 là tên của mặt hàng
• key_2 là “SoLuong” - value_2 là số lượng còn lại của mặt hàng
• key_3 là “GiaBan” - value_3 là giá của mặt hàng
Ví dụ: {“Ten” : “Bánh gạo”, “SoLuong” : 10, “GiaBan” : 30000}.
Bạn hãy viết chương trình thực hiện các yêu cầu sau:
• Nhập vào một danh sách các mặt hàng - ListMH , trong đó mỗi mặt hàng
bao gồm các thông tin tên mặt hàng, số lượng, giá bán theo cấu trúc từ điển
như trên. Kết thúc nhập khi tên mặt hàng là chuỗi rỗng.
• In ra các mặt hàng có số lượng còn dưới 5.
• In ra tổng số tiền hàng. Biết rằng tổng tiền hàng là tổng của (số lượng × giá
bán) của tất cả các mặt hàng.

```python
ds = []
dem = 1
def tongTien(ds):
    tong = 0
    for i in ds:
        tong += i['soLuong']*i['GiaBan']
    return tong



while True:
    print(f'---------\nnhập hang thu {dem}:')
    ten = input('nhập tên hang: ')
    if ten == '':break
    soLuong = int(input('nhập số lượng: '))
    giaBan = float(input('nhập giá bán: '))
    x = {'ten': ten, 'soLuong': soLuong,'GiaBan' : giaBan }
    ds.append(x)
    dem += 1
for i in ds:
    if i['soLuong'] < 5:
        print(f'-------\ntên hàng: {i['ten']} - số lượng: {i['soLuong']} - giá bán: {i['GiaBan']}')
tong = tongTien(ds)
print(f'************\ntổng tiền hàng là {tong}')

```

```python
import math as m
class Iris:
    def __init__(self, cdd = 0, crd = 0,cdc = 0 , crc = 0, kh = 'None' ):
        self.chieuDaiDai = cdd
        self.chieuRongDai = crd
        self.chieuDaiCanh = cdc
        self.chieuRongCanh = crc
        self.kieuHoa = kh
    def setKieuHoa(self,kh):
        self.kieuHoa = kh
    
    def khoangCach(self, bh):
        kq = 0
        tong = (self.chieuDaiDai - bh.chieuDaiDai)**2 + (self.chieuRongDai - bh.chieuRongDai)**2 + (self.chieuDaiCanh - bh.chieuDaiCanh)**2 + (self.chieuRongCanh - bh.chieuRongCanh)**2
        kq = round(m.sqrt(tong),2)
        return kq

x = []

tt = 1
while True:
    print(f'nhập bông hoa thứ {tt};')
    cdd = float(input('chiều dài đài: '))
    if cdd == 0 :break
    crd = float(input('chiều rộng đài: '))
    cdc = float(input('chiều dài cánh: '))
    crc = float(input('chiều rộng cánh: '))
    kh = input('kiểu hoa\'nếu chưa biết loại hoa thì điền\"None\"\': ')
    x.append(Iris(cdd,crd,cdc,crc,kh))
    tt += 1
thu = Iris(7.1,28.,6.4,1.6,'None')
kc = [(i.khoangCach(thu),i.kieuHoa) for i in x]
kc.sort()
luu = kc[:3]
dem = {}
for _ , i in luu :
    if i in dem :
        dem[i] += 1
    else:
        dem[i] = 1
max_loai = 0
loaiHoa = 'None'
for key , value in dem.items():
    if value > max_loai:
        max_loai = value
        loaiHoa = key

thu.setKieuHoa(loaiHoa)

print(thu.kieuHoa)


    
```
