Câu 3. (3.0 điểm): Xây dựng lớp Congnhan
mô tả thông tin về các công nhân gồm các thuộc tính: Mã công nhân, Họ tên, Bậc (từ 1
đến 3), số ngày làm việc, ngày ký hợp đồng ( thuộc tính ngày ký hợp đồng được thiết kế
lưu trữ theo kiểu từ điển gồm 3 khóa “day”, “month” và “year&quot; với giá trị tương ứng của
mỗi khóa là các số nguyên dương).
Viết các phương thức cho lớp Congnhan bao gồm:
- Phương thức khởi tạo giá trị các thuộc tính.
- Phương thức hiển thị thông tin về cán bộ bao gồm mã công nhân, họ tên, bậc, số
ngày làm việc, tiền lương. Biết rằng: Tiền lương = số ngày làm việc * tiền công nhật.
Trong đó tiền công nhật = 300000/ngày đối với bậc là 1, 350000/ngày đối với bậc 2 và
400000/ngày đối với bậc 3.
- Toán tử tải bội &gt; để so sánh hai đối tượng của lớp Congnhan (dựa vào tiêu chí
ngày ký hợp đồng).

```python
class Congnhan:
    def __init__(self, m = '',t = '',b = 0,snlv = 0,nkhd = {'day':0,'month':0,'year':0}):
        self.maCN = m
        self.ten = t
        self.bac = b
        self.soNgayLamViec = snlv
        self.ngayKyHopDong = nkhd
    def display(self):
        print('************************************************')
        print(f'thông tin cán bộ {self.ten}:')
        print(f'ma của công nhân: {self.maCN}')
        print(f'bậc của công nhân là {self.bac}')
        print(f'số ngày làm việc của công nhân là {self.soNgayLamViec}')
        # print(f'ngày ký hợp đồng với công nhân là {self.ngayKyHopDong}')
        tienLuong = 0
        if self.bac == 1:
            tienLuong = self.soNgayLamViec* 300000
        elif self.bac == 2:
            tienLuong = self.soNgayLamViec* 350000
        elif self.bac == 3:
            tienLuong = self.soNgayLamViec* 400000
        else :
            print('bậc không hợp lệ nên:')
        print(f'tiền lương công nhân nhận là {tienLuong}\n')
        
    def __gt__(self, cn):
        if self.ngayKyHopDong['year'] < cn.ngayKyHopDong['year']: return True
        if self.ngayKyHopDong['year'] == cn.ngayKyHopDong['year']: 
            if self.ngayKyHopDong['month'] < cn.ngayKyHopDong['month']: return True
            if self.ngayKyHopDong['month'] == cn.ngayKyHopDong['month']:
                if self.ngayKyHopDong['day'] < cn.ngayKyHopDong['day']:return True
        return False      


if __name__ == '__main__':
    thuTu = 1
    ds = []
    while 1:
        ss = 0
        print(f'nhập công nhân thứ {thuTu}:\n')
        ma = input('nhập mã công nhân: ')
        if ma == '':
            break
        for i in ds:
            if ma == i.maCN:
                ss += 1
                continue
        if ss != 0:
            print('nhập lại:')
            continue
        ten = input('nhập tên công nhân: ')
        bac = int(input('nhập bậc của công nhân(1<= bậc <= 3): '))
        snlv = int(input('nhập số ngày làm việc của công nhân: '))
        day , month , year = map(int , input('nhập ngày ký hợp đồng: ').split('/'))
        nkhd = {'day':day,'month':month ,'year' : year}
        cn = Congnhan(
                    ma,
                    ten,
                    bac,
                    snlv,
                    nkhd
                 )
        ds.append(cn)
        thuTu += 1
        
    for i in ds:
        i.display()
        
```

```python

class HOCPHAN:
    def __init__(self,ma = '',ten = '',sTC = 0,ngayd = {},mgv = ''):
        self.maHocPhan = ma
        self.tenDeCuong = ten
        self.sTinChi = sTC
        self.ngayDuyet = ngayd
        self.mGiaoVien = mgv
    def display(self):
        print(f"Mã học phần: {self.maHocPhan}")
        print(f"Tên đề cương: {self.tenDeCuong}")
        print(f"Số tín chỉ: {self.sTinChi}")
        if self.ngayDuyet:
            print(f"Ngày duyệt: {self.ngayDuyet['date']['day']}/{self.ngayDuyet['date']['month']}/{self.ngayDuyet['date']['year']} "
                  f"{self.ngayDuyet['time']['hour']}:{self.ngayDuyet['time']['minute']}")
        else:
            print("Ngày duyệt: Chưa được duyệt")
        print(f"Mã giảng viên: {self.mGiaoVien}")

class GIANGVIEN:
    def __init__(self,ma,ten):
        self.maGV = ma
        self.tenGV = ten
    def display(self):
        print(f"Mã giảng viên: {self.maGV}")
        print(f"Họ tên giảng viên: {self.tenGV}")
    

class DONVI:
    def __init__(self,ma,ten):
        self.maDV = ma
        self.tenDV = ten
    def display(self):
        print(f"Mã đơn vị: {self.maDV}")
        print(f"Tên đơn vị: {self.tenDV}")

class QUANLY:
    def __init__(self,dshp = [],dsgv = [],dsdv = []):
        self.__dshp = dshp
        self.__dsgv = dsgv
        self.__dsdv = dsdv
    def khoiTaoDuLieu(self,dshp,dsgv,dsdv):
        self.__dshp = dshp
        self.__dsgv = dsgv
        self.__dsdv = dsdv
    def display(self):
        print(f'{'Mã HP':<10} {'Tên đề cương':<35} {'Số TC':<10} {'Đơn vị phụ trách':<30} {'Người biên soạn':<25} {'Ngày duyệt'}')
        print()
        for i in self.__dshp:
            x = i.mGiaoVien[:3]
            gv = ''
            dv = ''
            for j in self.__dsgv:
                if i.mGiaoVien == j.maGV:
                    gv = j.tenGV
            for j in self.__dsdv:
                if x == j.maDV:
                    dv = j.tenDV
            print(f'{i.maHocPhan:<10} {i.tenDeCuong:<35} {i.sTinChi:<10} ',end='')
            print(f'{dv:<30} {gv:<25}',end='')
            if i.ngayDuyet:
                print(f"{i.ngayDuyet['date']['day']}/{i.ngayDuyet['date']['month']}/{i.ngayDuyet['date']['year']} "
                  f"{i.ngayDuyet['time']['hour']}:{i.ngayDuyet['time']['minute']}")
            else:
                print("Chưa được duyệt")

    def kiem(self):
        tuKhoa = input('nhập từ khóa để tìm kiếm: ')
        print('='*30)
        print('các mã học phần tìm kiếm được:')
        for i in self.__dshp:
            if tuKhoa.lower() in i.tenDeCuong.lower(): 
                print(i.maHocPhan,end='||')
        print()

    def thongKeDuyet(self):
        kq = 0
        for i in self.__dshp:
            if i.ngayDuyet:
                kq += 1
        print(f'số học phần đã được duyệt là : {kq}')
             
    def thongKeTungDV(self):
        print('='*30)
        for i in self.__dsdv:
            kq = 0
            for j in self.__dshp:
                x = j.mGiaoVien.strip()
                if x[:3] in i.maDV.strip():
                    kq += 1
            print(f'{i.tenDV:<30} {kq}')

    def thongKeTungGVMax(self):
        print('='*30)
        print('Danh sách các giảng viên phụ trách nhiều học phần nhất:')
        print(f'{'Tên giảng viên':<30} {'Số môn phụ trách'}')
        ds = {}
        for i in self.__dshp:
            if i.mGiaoVien in ds:
                ds[i.mGiaoVien] += 1
            else:
                ds[i.mGiaoVien] = 1
        maxtc = 0
        for _ , value in ds.items():
            if value > maxtc:
                maxtc = value
        for key , value in ds.items():
            if value == maxtc:
                for i in self.__dsgv:
                    if i.maGV.lower() == key.lower():
                        print(f'{i.tenGV:<30} {maxtc}')
    def donviHoanThanh(self):
        print('='*30)
        with open('donViHoanThanh.txt','w',encoding='utf-8') as f:
            ds1 = {}
            for i in self.__dshp:
                if i.ngayDuyet:
                    if i.maHocPhan[:3] in ds1:
                        ds1[i.maHocPhan[:3]] += 1
                    else:
                        ds1[i.maHocPhan[:3]] = 1
            ds2 = {}
            for i in self.__dshp:
                if i.mGiaoVien[:3] in ds2:
                    ds2[i.mGiaoVien[:3]] += 1
                else:
                    ds2[i.mGiaoVien[:3]] = 1
            for key,value in ds2.items():
                if key in ds1:
                    if ds1[key] == value:
                        f.write(f'{key} {value}\n')

   
def nhapTuFile():
    dshp = []
    dsgv = []
    dsdv = []
    with open('HP.txt','r',encoding='utf-8') as hp:
        for i in hp:
            ngayduyet = {}
            ma,ten,sotc,date,mGiaoVien = i.strip().split(',')
            sotc = int(sotc)
            if date :
                tach = date.split()
                ngay,thang, nam = map(int,tach[0].split('/'))
                gio, phut = map(int,tach[1].split(':'))
                ngayduyet = {'date':{'day':ngay,'month':thang,'year':nam},'time':{'hour':gio,'minute':phut}}
            hocphan = HOCPHAN(ma,ten,sotc,ngayduyet,mGiaoVien)
            dshp.append(hocphan)
    with open('GV.txt','r',encoding='utf-8') as gv:
        for i in gv:
            ma,ten = i.strip().split(',')
            dsgv.append(GIANGVIEN(ma,ten))
    with open('DV.txt','r',encoding='utf-8') as dv:
        for i in dv:
            ma,ten = i.strip().split(',')
            dsdv.append(DONVI(ma,ten))
    return QUANLY(dshp,dsgv,dsdv)

ql = nhapTuFile()
ql.display()
ql.kiem()
ql.thongKeDuyet()
ql.thongKeTungDV()
ql.thongKeTungGVMax()
ql.donviHoanThanh()
```

```python

import math as m
class Iris:
    def __init__(self, cdd = 0.0, crd = 0.0,cdc = 0.0 , crc = 0.0, kh = 'None' ):
        self.chieuDaiDai = cdd
        self.chieuRongDai = crd
        self.chieuDaiCanh = cdc
        self.chieuRongCanh = crc
        self.kieuHoa = kh
    def setKieuHoa(self,kh):
        self.kieuHoa = kh
    
    def khoangCach(self, bh):
        kq = 0
        tong = (self.chieuDaiDai - bh.chieuDaiDai)**2 + (self.chieuRongDai - bh.chieuRongDai)**2 + (self.chieuDaiCanh - bh.chieuDaiCanh)**2 + (self.chieuRongCanh - bh.chieuRongCanh)**2
        kq = round(m.sqrt(tong),2)
        return kq


def read_iris_data(filename):
    iris_list = []
    with open(filename, 'r') as file:
        for line in file:
            # Tách các giá trị trong dòng
            values = line.strip().split(',')
            x1 = float(values[0])
            x2 = float(values[1])
            x3 = float(values[2])
            x4 = float(values[3])
            kieu_hoa = values[4]
            iris_list.append(Iris(x1, x2, x3, x4, kieu_hoa))
    return iris_list
x = read_iris_data('iris.data')

tt = 1
thu = Iris(7.1,28.,6.4,1.6,'None')
kc = [(i.khoangCach(thu),i.kieuHoa) for i in x]
kc.sort()
# print(kc)
luu = kc[:3]
dem = {}
for _ , i in luu :
    if i in dem :
        dem[i] += 1
    else:
        dem[i] = 1
max_loai = 0
loaiHoa = 'None'
for key , value in dem.items():
    if value > max_loai:
        max_loai = value
        loaiHoa = key

thu.setKieuHoa(loaiHoa)

print(thu.kieuHoa)


    
```
